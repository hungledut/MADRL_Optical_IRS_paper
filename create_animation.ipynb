```python
import imageio
import glob
import re

frames = []


def frame_number(fname):
    return int(re.search(r"\d+", fname).group())

files = sorted(glob.glob("UAV_moving_*.png"), key=frame_number)


for file in files:
    frames.append(imageio.v2.imread(file))

imageio.mimsave("UAV_animation.gif", frames, duration=100)  # duration per frame in seconds
```
